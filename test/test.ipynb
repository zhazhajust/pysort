```python
from pysort import TrackParticles
```

```python
wkdir = "/home/yujq/users/caijie/Data/OsirisData/Undulator/gauss_laser/uniform_beam/w0_70/gamma10"
track = TrackParticles(wkdir)
times = track.times
```

```python
track.iter_all()
```

```python
x1 = track[0]
x3 = track[1]
p2 = track[2]
```

```python
from pypic import plot_lines, add_cax
import matplotlib.pyplot as plt
```

```python
keys = list(x1.keys())[::1000]
```

```python
fig, axs = plt.subplots(figsize = [4, 2])
for key in keys:
    line = plot_lines(axs, x1[key], x3[key], p2[key], cmap = "RdBu_r", clim = [-0.3, 0.3])

axs.set_xlim([0, 6000])
axs.set_ylim([-100, 100])
cax = add_cax(fig, axs)
cbar = plt.colorbar(line, cax)
```
